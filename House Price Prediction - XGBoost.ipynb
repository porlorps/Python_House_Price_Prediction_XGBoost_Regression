## Modeling

Due to our testing data does not contain target variable (SalePrice) which we will use it for prediction only, we split 20% of the training data into validation set for evaluation the Root Mean Square Error of the model.

```python
# Split training and validation data

def split_data(df, target_column, test_size=0.2, random_state=42):
    """
    Splits a DataFrame into training and testing sets.
    
    Parameters:
    df (DataFrame): The complete dataset containing features and target.
    target_column (str): The name of the target column.
    test_size (float): The proportion of the dataset to include in the test split.
    random_state (int): The seed used by the random number generator.
    
    Returns:
    X_train (DataFrame): Training features.
    X_test (DataFrame): Testing features.
    y_train (Series): Training target.
    y_test (Series): Testing target.
    """
    X = df.drop(columns=[target_column])  # All columns except the target
    y = df[target_column]                 # The target variable
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    return X_train, X_test, y_train, y_test

target_column = 'SalePrice'  # Set your target column name here
X1_train, X1_test, y1_train, y1_test = split_data(train_featured, target_column, test_size=0.2, random_state=42)
```

```python
# Training Regression Model

X_train = X1_train
X_test = X1_test
y_train = y1_train
y_test = y1_test

# Initialize and fit the XGBoost model
model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)  # Use 'reg:squarederror' for regression
model.fit(X_train, y_train)

```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
### Evaluation Matrix

def adjusted_r2_score(y_true, y_pred, n_features):
    """
    Calculate the adjusted R-squared score.
    
    Parameters:
    y_true (array-like): True target values.
    y_pred (array-like): Predicted target values.
    n_features (int): The number of features (independent variables) in the model.
    
    Returns:
    float: Adjusted R-squared score.
    """
    # Calculate R-squared
    r2 = r2_score(y_true, y_pred)
    
    # Number of observations
    n = len(y_true)
    
    # Calculate adjusted R-squared
    adjusted_r2 = 1 - (1 - r2) * ((n - 1) / (n - n_features - 1))
    
    return adjusted_r2


# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # Root Mean Squared Error
r2 = r2_score(y_test, y_pred)
n_features = X_test.shape[1]  # Number of features in the model
adj_r2 = adjusted_r2_score(y_test, y_pred, n_features)


print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R^2 Score: {r2:.2f}")
print("Adjusted R-squared:", adj_r2)
```

```python
# Feature Importance
importance = model.get_booster().get_score(importance_type='weight')
importance_df = pd.DataFrame(importance.items(), columns=['Feature', 'Importance'])
importance_df = importance_df.sort_values(by='Importance', ascending=False)
```

```python
# Plot the feature importance
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # Invert y-axis for descending order
plt.show()
```

```python
# Feature Importance
importance = model.get_booster().get_score(importance_type='weight')
importance_df = pd.DataFrame(importance.items(), columns=['Feature', 'Importance'])
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importance
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # Invert y-axis for descending order
plt.show()
```

When comparing the XGBoost model with feature selection importance above 100 to the model with all features, the larger model can produce more precision (less RMSE, higher R square) but the increasing in number of feature tend to have risk of overfitting when predict the unknow observations. Because our testing data does not contain actual house sales price, we cannot determine the different of two model on new predicted data, which model may be (or may be not) perform well if the number of features was increased. 

The adjusted r square of the model with more feature has incresed from 0.81 to 0.84. To reduce the chance of overfitting of the model, we then use the model with importance feature more than 100 for prediction.
